```python
# import dep
import os
import pandas as pd
import numpy as np
```

```python
# Create file path
file_path = os.path.join("Resources", "vine_table.csv")
```

```python
# load csv to df
df = pd.read_csv(file_path)
```

```python
df
```

```python
# count all reviews
total_review_count = df['review_id'].count()

total_review_count
```

```python
# filter for reviews with greater than 20 votes
df20 = df[df['total_votes'] > 20]

df20.reset_index(inplace=True)

df20
```

```python
# count of reviews with > 20 total votes
total20 = df20['review_id'].count()



total20
```

```python
# for loop to find % of helpful/total votes and add to df20
df20["percent"] = ""
for index in df20.index:
    index <= total20
    df20["percent"][index] = (df20["helpful_votes"][index] / df20["total_votes"][index])*100
print(f'complete at {index}')
```

```python
# filter for helpful percent >= 50
df50 = df20[df20["percent"] >= 50]

df50
```

```python
# create vine == y df
dfy = df50[df50["vine"] == "Y"]

dfy
```

```python
# count of vine sub reviews
total_count_of_vine = dfy['review_id'].count()

total_count_of_vine
```

```python
# create vine == n df
dfn = df50[df50["vine"] == "N"]

dfn
```

```python
# count of vine sub reviews
total_count_of_nonvine = dfn['review_id'].count()

total_count_of_nonvine
```

```python
# count of total reviews with 20+ votes and 50% helpful
total_helpful_reviews = df50['review_id'].count()

total_helpful_reviews
```

```python
# count of 5 star reviews with 20+ votes and 50% helpful

total_five_star_reviews = df50[df50['star_rating'] == 5]['star_rating'].count()

total_five_star_reviews
```

```python
# count for all vine == y with 5 star
total_y_five_review = dfy[dfy['star_rating'] == 5]['star_rating'].count()

total_y_five_review
```

```python
# calculate % of 5 star review that are vine subs
percent_y_review = round((total_y_five_review / (total_count_of_vine) * 100), 2)

percent_y_review
```

```python
# count for all vine == 5 with 5 star

total_n_five_review = dfn[dfn['star_rating'] == 5]['star_rating'].count()

total_n_five_review
```

```python
# # calculate % of 5 star review that are not vine subs
percent_n_review = round((total_n_five_review / total_count_of_nonvine) * 100, 2)

percent_n_review
```

### Summary Dataframes

```python
summary_df = pd.DataFrame({
    "total reviews": total_review_count,
    "total helpful reviews": total_helpful_reviews,
    "total vine subscribers": total_count_of_vine,
    "total non vine subscribers": total_count_of_nonvine}, index=[0])

summary_df
```

```python
summary_five_df = pd.DataFrame({
    "total 5 star reviews": total_five_star_reviews,
    "total 5 star vine subscribers reviews": total_y_five_review,
    "total 5 star non vine subscriber reviews": total_n_five_review,
}, index=[0])

summary_five_df
```

```python
percent_df = pd.DataFrame({
    "% 5 star vine subscriber reviews": percent_y_review,
    "% 5 star non vine subscriber reviews": percent_n_review
}, index=[0])

percent_df
```
